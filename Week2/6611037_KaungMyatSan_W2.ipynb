```python
df = pd.read_csv('W2_Superstore_Sales.csv',index_col=0)
df['Order_Date'] = pd.to_datetime(df['Order_Date'],errors = 'coerce')
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'],errors='coerce')
df['Order_Year'] = df['Order_Date'].dt.year
print(df.info())
```

```python
#Total Sales between 2015-2018
sales_by_year = df.groupby(['Order_Year'])['Sales'].sum()
print(sales_by_year)
total_sales = sales_by_year.sum()
print("Total Sales between 2015-2018:  " + str(total_sales))
```

```python
#Plot a bar chart for sales by Year
sales_by_year.plot(kind ='bar',color = 'black')
plt.title("Total Sales by Year (2015-2018)")
plt.xlabel("Order by Year")
plt.ylabel("Sales Volume")
plt.show()
```

```python
#Which product category has the highest sales amount(no. of order)
category_sales = df.groupby(['Category']).count()
category_sales.plot(kind ='bar',color = 'black')
plt.title('Orders by Product Category')
plt.xlabel('Category')
plt.ylabel('Number of Orders')
plt.show()
```

```python
#Sales volumes
sales_volume_by_category = df.groupby(['Category'])['Sales'].sum()
sales_volume_by_category.plot(kind ='bar',color = 'black')
plt.title("Total Sales Volume (2015-2018)")
plt.xlabel("Order by Year")
plt.ylabel("Sales Volume")
plt.grid(1)
plt.show()
```

```python
#Subcategory Sales
subcat = df.groupby(['Sub_Category'])['Sales'].sum().reset_index()
sorted_subcat = subcat.sort_values('Sales', ascending=False).reset_index(drop=True)
sorted_subcat.plot(kind ='bar',color = 'black')
plt.title("Total Subcategory (2015-2018)")
plt.xlabel("Subcategory")
plt.ylabel("Sales Volume")
plt.grid(1)
plt.show()
```

```python
#Sales by Region
region_sales = df.groupby('Region')['Sales'].sum()

sorted_regionSales = region_sales.sort_values(ascending=False)
sorted_regionSales.plot(kind='bar', color='black')
plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales (USD)')
plt.show()
```

```python
#Top 10 cities
city_sales = df.groupby('City')['Sales'].sum().reset_index()
city_sales = city_sales.sort_values(by = 'Sales', ascending = False)
top10cities_sales = city_sales.head(10)
top10cities_sales
```
