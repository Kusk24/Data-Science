```python
salaries = pd.read_csv('Salaries_with_date_of_response.csv', parse_dates = ['date_of_response'],date_format="%d/%m/%Y", index_col=0)
salaries 
```

```python
salaries.info()
```

```python
salaries['month'] = salaries['date_of_response'].dt.month
```

```python
salaries.drop(columns='Weekday', inplace=True)
```

```python
salaries['weekday'] = salaries['date_of_response'].dt.weekday
salaries
```

```python
sns.heatmap(salaries.corr(numeric_only=True), annot = True)
plt.show()
```

```python
twentyfifth = salaries['Salary_USD'].quantile(0.25)
median = salaries['Salary_USD'].median()
sevenfifth = salaries['Salary_USD'].quantile(0.75)
largest = salaries['Salary_USD'].max()
salary_labels = ['entry', 'mid', 'senior', 'exec']
ranges = [0,twentyfifth,median,sevenfifth,largest]

salaries['Salary_Level'] = pd.cut(salaries['Salary_USD'], labels = salary_labels, bins = ranges)

sns.countplot(data = salaries, x = 'Company_Size', hue = salaries['Salary_Level'])
plt.show()
```
